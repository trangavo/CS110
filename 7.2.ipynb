```python
# Q2

import random
import string
import time


def randomword(length):
    return ''.join(random.choice(string.ascii_lowercase) for i in range(length))


def empty_hash_table(N):
    return [[] for n in range(N)]


def add_to_hash_table(hash_table, item, hash_function):
    N = len(hash_table)
    # Your code here
    hash_table[hash_function%N].append(item)
    return hash_table


def contains(hash_table, item, hash_function):
    N = len(hash_table)
    # Your code here
    # return true if the item has already been stored in the hash_table
    if item in hash_table[hash_function%N]:
        return True

def remove(hash_table, item, hash_function):
    if not contains(hash_table, item, hash_function):
        raise ValueError()
    hash_table[hash_function%N].remove(item)
    return hash_table


def hash_str1(string):
    ans = 0
    for chr in string:
        ans += ord(chr)
    return ans


def hash_str2(string):
    ans = 0
    for chr in string:
        ans = ans ^ ord(chr)
    return ans


def hash_str3(string):
    ans = 0
    for chr in string:
        ans = ans * 128 + ord(chr)
    return ans


def hash_str4(string):
    random.seed(ord(string[0]))
    return random.getrandbits(32)

def Simulation1(N, num_words, length):
    hash_table = empty_hash_table(N)
    list_word = []
    for i in range(num_words):
        list_word.append(randomword(length))
    for item in list_word:
        add_to_hash_table(hash_table, item, hash_str1(item))
    # collisions
    collisions = 0
    for i in range(N):
        if len(hash_table[i]) > 1:
            collisions += len(hash_table[i]) - 1
    # averaged bucket size       
    buckets = 0
    for i in range(N):
        if len(hash_table[i]) > 0:
            buckets += 1
    bucket_size = num_words / buckets
    # finding
    start1 = time.time()
    found = 0
    not_found = 0
    for item in list_word:
        if contains(hash_table, item, hash_str1(item)):
            found += 1
        else:
            not_found += 1
    end1 = time.time()
    duration1 = end1 - start1
    # not stored
    start2 = time.time()
    not_in = []
    while len(not_in) < 10000:
        item = randomword(length)
        if contains(hash_table, item, hash_str1(item)) != True:
            not_in.append(item)
    
    end2 = time.time()
    duration2 = end2 - start2

    return "no of collisions:", collisions, "averaged bucket size:", bucket_size, "finding elements time", duration1, "finding not stored time", duration2

def Simulation2(N, num_words, length):
    hash_table = empty_hash_table(N)
    list_word = []
    for i in range(num_words):
        list_word.append(randomword(length))
    for item in list_word:
        add_to_hash_table(hash_table, item, hash_str2(item))
    # collisions
    collisions = 0
    for i in range(N):
        if len(hash_table[i]) > 1:
            collisions += len(hash_table[i]) - 1
    # averaged bucket size       
    buckets = 0
    for i in range(N):
        if len(hash_table[i]) > 0:
            buckets += 1
    bucket_size = num_words / buckets
    # finding
    start1 = time.time()
    found = 0
    not_found = 0
    for item in list_word:
        if contains(hash_table, item, hash_str2(item)):
            found += 1
        else:
            not_found += 1
    end1 = time.time()
    duration1 = end1 - start1
    # not stored
    start2 = time.time()
    not_in = []
    while len(not_in) < 10000:
        item = randomword(length)
        if contains(hash_table, item, hash_str2(item)) != True:
            not_in.append(item)
    
    end2 = time.time()
    duration2 = end2 - start2

    return "no of collisions:", collisions, "averaged bucket size:", bucket_size, "finding elements time", duration1, "finding not stored time", duration2


def Simulation3(N, num_words, length):
    hash_table = empty_hash_table(N)
    list_word = []
    for i in range(num_words):
        list_word.append(randomword(length))
    for item in list_word:
        add_to_hash_table(hash_table, item, hash_str3(item))
    # collisions
    collisions = 0
    for i in range(N):
        if len(hash_table[i]) > 1:
            collisions += len(hash_table[i]) - 1
    # averaged bucket size       
    buckets = 0
    for i in range(N):
        if len(hash_table[i]) > 0:
            buckets += 1
    bucket_size = num_words / buckets
    # finding
    start1 = time.time()
    found = 0
    not_found = 0
    for item in list_word:
        if contains(hash_table, item, hash_str3(item)):
            found += 1
        else:
            not_found += 1
    end1 = time.time()
    duration1 = end1 - start1
    # not stored
    start2 = time.time()
    not_in = []
    while len(not_in) < 10000:
        item = randomword(length)
        if contains(hash_table, item, hash_str3(item)) != True:
            not_in.append(item)
    
    end2 = time.time()
    duration2 = end2 - start2

    return "no of collisions:", collisions, "averaged bucket size:", bucket_size, "finding elements time", duration1, "finding not stored time", duration2


def Simulation4(N, num_words, length):
    hash_table = empty_hash_table(N)
    list_word = []
    for i in range(num_words):
        list_word.append(randomword(length))
    for item in list_word:
        add_to_hash_table(hash_table, item, hash_str4(item))
    # collisions
    collisions = 0
    for i in range(N):
        if len(hash_table[i]) > 1:
            collisions += len(hash_table[i]) - 1
    # averaged bucket size       
    buckets = 0
    for i in range(N):
        if len(hash_table[i]) > 0:
            buckets += 1
    bucket_size = num_words / buckets
    # finding
    start1 = time.time()
    found = 0
    not_found = 0
    for item in list_word:
        if contains(hash_table, item, hash_str4(item)):
            found += 1
        else:
            not_found += 1
    end1 = time.time()
    duration1 = end1 - start1
    # not stored
    start2 = time.time()
    not_in = []
    while len(not_in) < 10000:
        item = randomword(length)
        if contains(hash_table, item, hash_str4(item)) != True:
            not_in.append(item)
    
    end2 = time.time()
    duration2 = end2 - start2

    return "no of collisions:", collisions, "averaged bucket size:", bucket_size, "finding elements time", duration1, "finding not stored time", duration2

for i in range(5):
    print(i, "time")
    print("hash_function 1", Simulation1(5000, 100000, 10))
    print("hash_function 2", Simulation2(5000, 100000, 10))
    print("hash_function 3", Simulation3(5000, 100000, 10))
    print("hash_function 4", Simulation4(5000, 100000, 10))

```
